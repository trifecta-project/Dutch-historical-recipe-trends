<h1>Cookbooks

```python
import pandas as pd
import glob

# Cookbooks dataset
csv_files = ('1910_count.csv', '1912_count.csv', '1925_count.csv','1932_count.csv', '1938_count.csv','1939_count.csv', '1940_count.csv')

normalized_data = {}

# Defining ingredient synonyms
ingredient_mapping = {
    'aardappel': ['aardappel','aardappelblokjes','aardappelen','aardappelpartjes','aardappels','aardappelschijfjes'],
    'abrikoos':['abrikozen', 'abrikoos'],
    'amandel':['amandel','amandelen'],
    'ansjovis':['ansjovis','ansjovissen'],
    'appel':['appel','appelen','appels'],
    'banaan':['banaan', 'bananen'],
    'brood': ['brood', 'broodje'],
    'cherry tomaten':['cherry tomaten','cherry tomaatjes', 'cherrytomaten'],
    'chilli':['chilli', 'shillipeper'],
    'citroen':['citroen','citroenen','citroensap'],
    'courgette': ['courgette', 'courgettes'],
    'ei': ['eieren', 'ei'],
    'eiwit': ['witte', 'eiwit'],
    'kaas': ['geraspte', 'kaas'],
    'margarine': ['boter of margarine'],
    'ui':['ui','uien']
}

for file in csv_files:
    decade = file.split("_")[-2] + "-" + file.split("_")[-1].split(".")[0]

    df = pd.read_csv(file)

    df = df.dropna(subset=['Ingredients'])

    total_recipes = len(df)

    ingredient_counts = {}
    for ingredients in df["Ingredients"]:
        ingredient_list = [i.strip() for i in ingredients.split(",")]
        for ingredient in ingredient_list:
            for key, synonyms in ingredient_mapping.items():
                if ingredient in synonyms:
                    ingredient = key 
                    break 
            
            ingredient_counts[ingredient] = ingredient_counts.get(ingredient, 0) + 1

    # Relative frequency calculation
    normalized_counts = {k: v / total_recipes for k, v in ingredient_counts.items()}

    normalized_data[decade] = normalized_counts

normalized_df = pd.DataFrame(normalized_data).fillna(0)

normalized_df.to_csv("relative_ingredient_frequency_cookbooks.csv")

print("Results saved")
```
